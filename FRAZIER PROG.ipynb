```python
# this only used when u use regression analysis i.e linear and decision tree regression.
import pandas as pd 
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
import matplotlib.pyplot as plt
import numpy as np
```

```python
# decleare columns you want to use as features
features = ["Rating","Metascore"]
```

```python
# Specify the Prediction Target 
target = ["Revenue (Millions)"]
```

```python
#Extract Features and Target ("Revenue (Millions)") Values into Separate Dataframes 
X = i[features]
y = i[target]
```

```python
# Let us look at a typical row from our features:
X.iloc[2]
```

```python
X # for features only ie ratings and metascore
```

```python
y # for target only ie revenue
```

```python
# Split the Dataset into Training and Test Datasets 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=324)
```

```python
#(1) Linear Regression: Fit a model to the training set 
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Perform Prediction using Linear Regression Model (remaining test sizei.e(0.33) 
y_prediction = regressor.predict(X_test)
y_prediction
```

```python
y_test.describe()# describing target
```

```python
#Evaluate Linear Regression Accuracy using Root Mean Square Error 
RMSE = sqrt(mean_squared_error(y_true = y_test, y_pred = y_prediction))
```

```python
RMSE
```

```python
#(2) Decision Tree Regressor: Fit a new regression model to the training set 
regressor = DecisionTreeRegressor(max_depth=20)
regressor.fit(X_train, y_train)
```

```python
#Perform Prediction using Decision Tree Regressor 
y_prediction = regressor.predict(X_test)
y_prediction
```

```python
#For comparision: What is the mean of the expected target value in test set ? 
y_test.describe()
```

```python
#Evaluate Decision Tree Regression Accuracy using Root Mean Square Error 
RMSE = sqrt(mean_squared_error(y_true = y_test, y_pred = y_prediction))
```

```python
RMSE
```
